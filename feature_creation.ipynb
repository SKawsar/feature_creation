```python
import pandas as pd
import numpy as np
```

```python
# sample dataframe
df = pd.DataFrame({"col_1": [0,5,10,15,20],
                   "col_2": [1,2,3,4,5]})

display(df)
```

## Addition

```python
df["addition"] = df["col_1"] + df["col_2"]
display(df)
```

## Subtraction

```python
df["subtraction"] = df["col_1"] - df["col_2"]
display(df)
```

## Absolute

```python
df["absolute"] = np.abs(df["subtraction"])
display(df)
```

## Multiplication

```python
df["multiplication"] = df["col_1"] * df["col_2"]
display(df)
```

## Division

```python
df["division"] = df["col_1"] / df["col_2"]
display(df)
```

## Floor

```python
df["floor"] = df["col_1"] // df["col_2"]
display(df)
```

## Ceiling

```python
df["ceil"] = np.ceil(df["col_1"] / df["col_2"])
display(df)
```

## Remainder

```python
df["remainder"] = df["col_1"] % df["col_2"]
display(df)
```

## Square

```python
df["square"] = df["col_1"]**2
display(df)
```

## Square Root

```python
df["square_root"] = np.sqrt(df["col_1"])
display(df)
```

## Cube

```python
df["cube"] = df["col_2"]**3
display(df)
```

## Exponent (n-th power)

```python
df["exponent"] = df["col_1"]**df["col_2"]
display(df)
```

```python
# sample dataframe
df = pd.DataFrame({"col_1": [1,2,3,4,5]})

display(df)
```

## Exponential

```python
df["exponential"] = np.exp(df["col_1"])
display(df)
```

## Natural Logarithm, log base 2, log base 10

```python
df["log"] = np.log(df["col_1"])
df["log2"] = np.log2(df["col_1"])
df["log10"] = np.log10(df["col_1"])
display(df)
```

## Standardization or z-score method

```python
df["standardized"] = (df["col_1"] - df["col_1"].mean())/df["col_1"].std()
display(df)
```

## Normalization

```python
df["normalized"] = (df["col_1"] - df["col_1"].min()) / (df["col_1"].max() - df["col_1"].min()) 
display(df)
```

```python
# sample dataframe
df = pd.DataFrame({"col_1": [0,5,10,15,20],
                   "col_2": [1,2,3,4,5]})

display(df)
```

## Lag

```python
df["lag"] = df["col_1"].shift(2)
display(df)
```

## Lead

```python
# lead
df["lead"] = df["col_1"].shift(-2)
display(df)
```

```python
# sample dataframe
df = pd.DataFrame({"radian": [0, np.pi/6, np.pi/4, np.pi/3]})

display(df)
```

## Radian to Degree conversion

```python
df["degree"] = np.rad2deg(df["radian"])
display(df)
```

## Degree to Radian conversion

```python
df["radian_converted"] = np.deg2rad(df["degree"])
display(df)
```

## Trigonometric functions: sinx, cosx, tanx, sinhx, coshx, tanhx

```python
df["sinx"] = np.sin(df["radian"])
df["cosx"] = np.cos(df["radian"])
df["tanx"] = np.tan(df["radian"])
df["sinhx"] = np.sinh(df["radian"])
df["coshx"] = np.cosh(df["radian"])
df["tanhx"] = np.tanh(df["radian"])

display(df)
```

## Round

```python
# rounding up to 3 decimal points
df["sinx"] = np.round(np.sin(df["radian"]), 3)

# rounding up to 2 decimal points
df["cosx"] = np.round(np.cos(df["radian"]), 2)

# rounding up to 4 decimal points 
df["tanx"] = np.round(np.tan(df["radian"]), 4)

display(df)
```

```python
df = pd.DataFrame({"Date" : ["1/1/2022", "1/2/2022", "1/3/2022", "1/4/2022", "1/5/2022"],
                   "Event" : ["Music", "Poetry", "Music", "Music", "Poetry"]})

display(df)
```

### Creating a new column based on a given condition over the existing column using List comprehension, if else and for loop

Example 1: Create a new column "Price" and set Price = 200 if the "Event" is "Music" and Price = 100 if the "Event" is "Poetry"

```python
# method 1
df["Price_1"] = [200 if x =='Music' else 100 for x in df["Event"]]

# method 2
df["Price_2"] = np.where(df["Event"] == "Music", 200, 100)

# method_3
df["Price_3"] = df["Event"].map({"Music": 200, "Poetry": 100})

display(df)
```

### How to concat multiple columns in string format

```python
df["Decision"] = df["Event"] + " event ticket price is $" + df["Price_1"].astype(str)

display(df)
```
